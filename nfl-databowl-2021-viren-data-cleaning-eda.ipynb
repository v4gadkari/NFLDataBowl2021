```python
gamePlaysSubset = ['quarter', 'down', 'yardsToGo', 'playType', 'yardlineNumber', 'offenseFormation', 'defendersInTheBox', 'numberOfPassRushers', 'typeDropback', 'absoluteYardlineNumber', 'passResult']


game_plays_combined_model_features = game_plays_combined_features_df[gamePlaysSubset]


game_plays_combined_model_features
```

```python
game_plays_combined_model_features['playType'].value_counts()

game_plays_combined_model_features['playType'] = game_plays_combined_model_features['playType'].apply(lambda row: 0 if row == 'play_type_unknown' else 1 if row == 'play_type_sack' else 2)

game_plays_combined_model_features
```

```python
game_plays_combined_model_features['offenseFormation'].value_counts()


game_plays_combined_model_features['offenseFormation'] = game_plays_combined_model_features['offenseFormation'].apply(lambda row: 0 if row == 'WILDCAT' else 1 if row == 'JUMBO' else 2 if row == 'PISTOL' else 3 if row == 'I_FORM' else 4 if row == 'EMPTY' else 5 if row == 'SINGLEBACK' else 6)

game_plays_combined_model_features
```

```python
game_plays_combined_model_features['typeDropback'].value_counts()

game_plays_combined_model_features['typeDropback'] = game_plays_combined_model_features['typeDropback'].apply(lambda row: 0 if row == 'UNKNOWN' else 1 if row == 'DESIGNED_ROLLOUT_LEFT' else 2 if row == 'SCRAMBLE_ROLLOUT_LEFT' else 3 if row == 'DESIGNED_ROLLOUT_RIGHT' else 4 if row == 'SCRAMBLE' else 5 if row == 'SCRAMBLE_ROLLOUT_RIGHT' else 6)

game_plays_combined_model_features
```

```python
subset = ['quarter', 'down', 'yardsToGo', 'playType', 'yardlineNumber', 'offenseFormation', 'defendersInTheBox', 'numberOfPassRushers', 'typeDropback', 'absoluteYardlineNumber']

game_plays_model_features_final = game_plays_combined_model_features[subset]

game_plays_model_features_final
```
